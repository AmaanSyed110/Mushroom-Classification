## **The aim of this project is to build a classification model that classifies the type of mushroom into it's correct class**

```python
#import the necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
```

```python
#mounting google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#importing dataset from google drive
df=pd.read_csv('/content/drive/MyDrive/Datasets/mushrooms.csv')
```

```python
#displaying top 5 values from dataset
df.head()
```

```python
#information about the dataset
df.info()
```

```python
#statistical description of the dataset
df.describe()
```

```python
#checking for null values in dataset
df.isna().sum()
```

```python
#converting the categorical values to numeric values for evaluation
le=LabelEncoder()
for columns in df.columns:
  df[columns]=le.fit_transform(df[columns])
```

```python
df.head()
```

```python
#selecting the feature and target variable
X=df.drop(['class'],axis=1)
y=df['class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#SVC MODEL
model=SVC()
model.fit(X_train,y_train)
```

```python
#predictions on SVC Model
pred=model.predict(X_test)
```

```python
#accuracy of the model
cr=classification_report(y_test,pred)
print(cr)
```

```python
#Random Forest Classifier Model
rfc_model=RandomForestClassifier(n_estimators=200)
rfc_model.fit(X_train,y_train)
```

```python
#Predictions on RFC Model
rfc_pred=rfc_model.predict(X_test)
```

```python
#Accuracy of RFC Model
rfc_cr=classification_report(y_test,rfc_pred)
print(rfc_cr)
```

```python
#Decision Tree Classifier Model
dtc_model=DecisionTreeClassifier()
dtc_model.fit(X_train,y_train)
```

```python
#predictions of model
dtc_pred=dtc_model.predict(X_test)
```

```python
#accuracy of model
dtc_cr=classification_report(y_test,dtc_pred)
print(dtc_cr)
```
